```python
def extract_schedule_info(sentence):
    # 날짜와 시간을 추출하는 정규표현식
    date_time_pattern = r'(\d+월) (\d+일) (오전|오후) (\d+)시(?: (\d+)분)?'
    match = re.search(date_time_pattern, sentence)
    if match:
        month = match.group(1).strip("월")
        day = match.group(2).strip("일")
        time_period = match.group(3)
        hour = match.group(4)
        if time_period == '오후':
            num_hour = hour.strip("시")
            hour = str(int(hour) + 12)# 오후면 시간에 12를 더해줍니다.
        minute = match.group(5) if match.group(5) else "0"  # 분 정보가 없으면 0으로 처리합니다.
        todo = sentence[match.end():].strip()  # 할 일은 날짜와 시간 이후의 부분입니다.
        return month, day, hour, minute, todo
    else:
        return None
```

```python
schedule_sentences = [
    "3월 15일 오후 2시에 친구와 카페에서 공부 모임이 있어요.",
    "4월 10일 오전 11시에 할머니 집에서 가족들과 점심 식사를 할 예정입니다."
    ,"4월 2일 오후 1시, 놀이공원.", "5월 28일 오후 3시 15분에 친구와 커피를 마시러 갈 예정입니다."
]

# 각 일정을 추출하고 리스트에 저장합니다.
schedules = []
for sentence in schedule_sentences:
    schedule_info = extract_schedule_info(sentence)
    if schedule_info:
        schedules.append(schedule_info)

# 엑셀 파일에 쓰기 위한 코드
import pandas as pd

# 추출된 일정 정보를 데이터프레임으로 변환
df = pd.DataFrame(schedules, columns=['Month', 'Day', 'Hour', 'Minute', 'ToDo'])
```
